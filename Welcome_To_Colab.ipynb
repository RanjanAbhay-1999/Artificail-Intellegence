```
# Importing dependencies
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```
# Years and corresponding global temperature anomalies
years = np.array([1880, 1890, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2030])
temperature_anomalies = np.array([-0.12, -0.08, -0.05, -0.02, 0.03, 0.08, 0.13, 0.18, 0.23, 0.28, 0.33, 0.38, 0.43, 0.48, 0.53, 0.58])

# Reshape the data for inear regression
X = years.reshape(-1, 1)
y = temperature_anomalies

# Create a linear regession model
model = LinearRegression()

# Train the model to fit
model.fit(X, y)

def main():
  print("Climate Change Prediction")

  year = int(input("Enter a year for temperature prediction: "))

  # Predict the temperature anomaly using the trained model
  predicted_anomaly = model.predict([[year]])

  print(f"Predicted temperature anomaly for {year}: {predicted_anomaly[0]:.2f}°C")

  # Visualize the data and regession line
  plt.scatter(years, temperature_anomalies, color='blue', label='Actual Data')
  plt.xlabel("Year")
  plt.ylabel("Temperature Anomaly (°C)")
  plt.title("Global Temperature Anomalies Over Time")
  plt.legend()
  plt.show()
if __name__=="__main__":
  main()
```
